### 3. 모델링

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error

from xgboost import XGBRegressor
from lightgbm import LGBMRegressor
from sklearn.ensemble import GradientBoostingRegressor, RandomForestRegressor
```

#### 3-1. rmse 함수 만들기

```python
def rmse(y_test, y_pred):
    return np.sqrt(mean_squared_error(np.expm1(y_test), np.expm1(y_pred)))
```

#### 3-2. model 객체 생성

```python
random_state = 2020

gboost = GradientBoostingRegressor(random_state=random_state)
xgboost = XGBRegressor(random_state=random_state)
lightgbm = LGBMRegressor(random_state=random_state)
rdforest = RandomForestRegressor(random_state=random_state)

models = [gboost, xgboost, lightgbm, rdforest]
```

#### 3-3. model score 가져오는 함수 생성

```python
def get_scores(models, train, y):
    df = {}
    
    for model in models:
        model_name = model.__class__.__name__
        
        X_train, X_test, y_train, y_test = train_test_split(train,y,
                                                           random_state = random_state,
                                                           test_size=0.2)
        
        # 모델에 학습시킨다
        model.fit(X_train, y_train)
        
        # 예측하기
        y_pred = model.predict(X_test)
        
        df[model_name]= rmse(y_test, y_pred)
        
        score_df = pd.DataFrame(df, index=['RMSE']).T.sort_values('RMSE', ascending=False)
        
    return score_df

get_scores(models, train, y)
```

### 4. 모델 튜닝하기

```python
from sklearn.model_selection import GridSearchCV
```

```python
def my_GridSearch(model, train, y, param_grid, verbose=2, n_jobs=5):
    # 1. GridSearchCV 모델로 model 초기화
    grid_model = GridSearchCV(model, param_grid=param_grid,
                             scoring = 'neg_mean_squared_error',
                             cv=5, verbose=2, n_jobs=5)
    # 2. 모델 학습
    grid_model.fit(train, y)
    
    # 3. 결과 저장
    params = grid_model.cv_results_['params']
    scores = grid_model.cv_results_['mean_test_score']
    
    # 4. 데이터프레임 생성 후 RMSLE 값을 추가하여 점수높은 순 result 반환
    results = pd.DataFrame(params)
    results['score'] = scores
    results['RMSLE'] = np.sqrt(-1 * results['score'])
    results.sort_values('RMSLE', ascending=False, inplace=True)
    
    return results
```

#### 4-1. GradientBoostingRegressor

```python
param_grid_gb = {
    'n_estimators': [200,300,400],
    'max_depth': [4, 6, 8],
    'learning_rate': [0.1],
    'subsample' : [0.7], 
}
```

```python
my_GridSearch(gboost, train, y, param_grid_gb, verbose=2, n_jobs=5)
```

#### 4-5. 모델 성능 평가 및 average blending

```python
from sklearn.model_selection import KFold, cross_val_score

models = [{'model':gboost, 'name':'GradientBoosting'}, {'model':rdforest, 'name':'RandomForest'},
          {'model':lightgbm, 'name':'LightGBM'}, {'model':xgboost, 'name':'XGBoost'}]

def get_cv_score(models):
    kfold = KFold(n_splits=5, random_state=2020, shuffle=True).get_n_splits(train.values)
    for m in models:
        print("Model {} CV score : {:.4f}".format(m['name'], np.mean(cross_val_score(m['model'], train.values, y)), 
                                             kf=kfold))
        
get_cv_score(models)
```

```python
def AveragingBlending(models, x, y, sub_x):
    for m in models : 
        m['model'].fit(x.values, y)
    
    predictions = np.column_stack([
        m['model'].predict(sub_x.values) for m in models
    ])
    return np.mean(predictions, axis=1)
```

```python
y_pred = AveragingBlending(models, train, y, test)
```

```python
print(len(y_pred))
y_pred
y_pred = np.expm1(y_pred)
y_pred
```
