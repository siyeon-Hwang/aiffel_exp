# 1. 손글씨 분류하기

```python
#1.필요한 모듈 호출하기
import pandas as pd
from sklearn.datasets import load_digits
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
```

```python
#2.데이터 준비
digits = load_digits()
```

```python
#3.데이터 이해하기
digits_data = digits.data
digits_label = digits.target
print(digits_data.shape)
print(digits_label.shape)
print('Target Names: ',set(digits_label))
print(digits.DESCR)
```

```python
digits_df = pd.DataFrame(data = digits_data, columns = digits.feature_names)
#digits_df.head()

digits_df['label'] = digits.target
digits_df.head()
```

```python
#4.train-test 데이터 분리
X_train, X_test, y_train, y_test = train_test_split(digits_data,
                                                   digits_label,
                                                   test_size=0.2,
                                                   random_state=7)

print(X_train.shape, y_train.shape)
print(X_test.shape, y_train.shape)
```

```python
#5.다양한 모델로 학습시켜보기
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn import svm
from sklearn.linear_model import SGDClassifier
from sklearn.linear_model import LogisticRegression

#객체생성
tree = DecisionTreeClassifier()
rf_tree = RandomForestClassifier()
svm_model = svm.SVC()
sgd_model = SGDClassifier()
lr_model = LogisticRegression(max_iter = 10000, solver = 'lbfgs')

#모델에 학습시키기
tree.fit(X_train, y_train)
rf_tree.fit(X_train, y_train)
svm_model.fit(X_train, y_train)
sgd_model.fit(X_train, y_train)
lr_model.fit(X_train, y_train)

#y 값 예측하기
y_tree = tree.predict(X_test)
y_rf = rf_tree.predict(X_test)
y_svm = svm_model.predict(X_test)
y_sgd = sgd_model.predict(X_test)
y_lr = lr_model.predict(X_test)
```

```python
#6.모델 평가해보기
#print('#1.DecisionTree metrics: \n', classification_report(y_test, y_tree))
#print('#2.RandomForest metrics: \n', classification_report(y_test, y_rf))
print('#3.SVM metrics: \n', classification_report(y_test, y_svm))
#print('#4.SGDClassifer metrics: \n', classification_report(y_test, y_sgd))
#print('#5.LogisticRegression metrics: \n', classification_report(y_test, y_lr))
```

숫자데이터를 분류하는 데이터로는 모델의 성능을 평가하는 지표로 정확도(accuracy)를 보는 게 좋을 것 같다. 위에서 사용한 숫자데이터 라벨들의 크기들이 엄청나게 차이나는게 아니기 때문에 불균형 데이터에 이용하는 f1 score를 굳이 이용하지 않아도 될 것 같다. 

# 2. 와인 분류하기

```python
#1.필요모듈 호출
from sklearn.datasets import load_wine
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
import pandas as pd
```

```python
#2.데이터 준비
wine = load_wine()
```

```python
#3.데이터 이해
wine_data = wine.data
wine_label = wine.target

print(wine_data.shape, wine_label.shape)
print(wine.target_names)
print(wine.DESCR)
```

```python
wine_df = pd.DataFrame(data=wine.data, columns=wine.feature_names)
wine_df['label'] = wine.target

wine_df.tail()
```

```python
#4.train-test split
X_train, X_test, y_train, y_test = train_test_split(wine_data,
                                                   wine_label,
                                                   test_size=0.2,
                                                   random_state=32)

print(X_train.shape, y_train.shape)
print(X_test.shape, y_test.shape)
```

```python
#5.다양한 모델에 학습시키기
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn import svm
from sklearn.linear_model import SGDClassifier
from sklearn.linear_model import LogisticRegression

#객체생성
tree = DecisionTreeClassifier()
rf_tree = RandomForestClassifier()
svm_model = svm.SVC()
sgd_model = SGDClassifier()
lr_model = LogisticRegression(max_iter = 10000, solver = 'lbfgs')

#모델에 학습시키기
tree.fit(X_train, y_train)
rf_tree.fit(X_train, y_train)
svm_model.fit(X_train, y_train)
sgd_model.fit(X_train, y_train)
lr_model.fit(X_train, y_train)
```

```python
#6.모델 평가하기_정확도 보기
#print('#1.DecisionTree train Score: {:.3f}'.format(tree.score(X_train, y_train)))
print('#1.DecisionTree Score: {:.3f}'.format(tree.score(X_test, y_test)))
print('#2.RandomForest Score: {:.3f}'.format(rf_tree.score(X_test, y_test)))
#print('#3.  SVM train Score : {:.3f}'.format(svm_model.score(X_train, y_train)))
print('#3.     SVM Score    : {:.3f}'.format(svm_model.score(X_test, y_test)))
print('#4.SGDClassifier Score: {:.3f}'.format(sgd_model.score(X_test, y_test)))
#print('#5.LogisticRegression train Score: {:.3f}'.format(lr_model.score(X_train, y_train)))
print('#5.LogisticRegression Score: {:.3f}'.format(lr_model.score(X_test, y_test)))
```

```python
#confusion matrix로 분류가 잘 되었나 확인
from sklearn.metrics import confusion_matrix

y_rf = rf_tree.predict(X_test)
y_svm = svm_model.predict(X_test)

#print(classification_report(y_test, y_rf))
confusion_rf = confusion_matrix(y_test, y_rf)
confusion_svm = confusion_matrix(y_test, y_svm)

print('#2.RandomForest confusion matrix: \n',confusion_rf)
print('#3.SVM confusion matrix: \n', confusion_svm)
```

숫자데이터와 달리 와인데이터는 모델 간 정확도에 꽤 많은 차이가 난다. 가장 좋은 성능을 보인 모델은 DecisionTree와 RandomForest 이고, LogisticRegression의 성능도 꽤 높았다.<br> 이를 평가하기 적절한 지표를 고르라면 숫자데이터와 마찬가지로 정확도를 꼽을 것이다.<br>
클래스 간 데이터의 개수가 크게 차이나는 것도 아니고, 단순히 숫자를 분류하고자 하기 때문이다.

# 3. 유방암 진단하기

```python
#1.필요한 모듈 호출
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
import pandas as pd
```

```python
#2.데이터 준비
breast_cancer = load_breast_cancer()
```

```python
#3.데이터 이해
cancer_data = breast_cancer.data
cancer_label = breast_cancer.target

print(cancer_data.shape, cancer_label.shape)
print(breast_cancer.target_names)
print(breast_cancer.DESCR)
```

```python
cancer_df = pd.DataFrame(data=cancer_data, columns = breast_cancer.feature_names)
cancer_df['label'] = breast_cancer.target
cancer_df.tail()
```

```python
#4.train-test split
X_train, X_test, y_train, y_test = train_test_split(cancer_data,
                                                   cancer_label,
                                                   test_size=0.2,
                                                   random_state=32)

print(X_train.shape, y_train.shape)
print(X_test.shape, y_test.shape)
```

```python
#5.다양한 모델에 학습시키기
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn import svm
from sklearn.linear_model import SGDClassifier
from sklearn.linear_model import LogisticRegression

#객체생성
tree = DecisionTreeClassifier()
rf_tree = RandomForestClassifier()
svm_model = svm.SVC()
sgd_model = SGDClassifier()
lr_model = LogisticRegression(max_iter = 10000, solver = 'lbfgs')

#모델에 학습시키기
tree.fit(X_train, y_train)
rf_tree.fit(X_train, y_train)
svm_model.fit(X_train, y_train)
sgd_model.fit(X_train, y_train)
lr_model.fit(X_train, y_train)

#y 값 예측하기
y_tree = tree.predict(X_test)
y_rf = rf_tree.predict(X_test)
y_svm = svm_model.predict(X_test)
y_sgd = sgd_model.predict(X_test)
y_lr = lr_model.predict(X_test)
```

```python
#6.모델평가
#print('#1.DecisionTree metrics: \n', classification_report(y_test, y_tree))
print('#2.RandomForest metrics: \n', classification_report(y_test, y_rf))
#print('#3.SVM metrics: \n', classification_report(y_test, y_svm))
print('#4.SGDClassifer metrics: \n', classification_report(y_test, y_sgd))
#print('#5.LogisticRegression metrics: \n', classification_report(y_test, y_lr))
```

위의 metrics를 보면, RandomForest가 정확도 94%로 가장 높고, SGDClassifier가 정확도 77%로 가장 낮게 나왔다. <br>
하지만 암 관련 데이터 특성 상 정확도를 따지기보다는, 실제로 악성 종양인데 일반종양으로 분류할 오류를 잡는 것이 더 중요하기 때문에 recall 지표와 f1 score를 따지는 것이 좋다고 생각한다. <br>
이를 확인해보니 마찬가지로 RandomForest가 가장 높은 성능을 보인다.<br> SGDClassifier의 경우 1(일반종양)에 대한 recall지표는 높게 나왔지만, 0(악성종양)에 대한 지표는 형편없다. <br>
아래의 confusion_matrix로 확인해보자

```python
from sklearn.metrics import confusion_matrix

y_rf = rf_tree.predict(X_test)
y_sgd = sgd_model.predict(X_test)

#print(classification_report(y_test, y_rf))
confusion_rf = confusion_matrix(y_test, y_rf)
confusion_sgd = confusion_matrix(y_test, y_sgd)

print('#2.RandomForest confusion matrix: \n',confusion_rf)
print('#3.SVM confusion matrix: \n', confusion_sgd)
```

SVM confusion matrix를 보면 실제로 일반종양(1)인 데이터는 한치의 오차 없이 잘 맞췄지만, 실제로 악성종양(0)인 데이터는 맞힌 것보다 틀린 것이 많다. <br>
