## 1. 인물모드 사진 만들기

### 1-1. 인물사진

#### 1) 이미지 가져오기

```python
import cv2
import numpy as np
from pixellib.semantic import semantic_segmentation
from matplotlib import pyplot as plt
```

```python
img_path1 = '../data/human_segmentation/images/jun3.jpg'
img_orig1 = cv2.imread(img_path1)

print(img_orig1.shape)
plt.imshow(cv2.cvtColor(img_orig1, cv2.COLOR_BGR2RGB))
plt.show()
```

#### 2) 모델 불러오기

```python
model_file = '../data/human_segmentation/models/deeplabv3_xception_tf_dim_ordering_tf_kernels.h5'

#모델 생성
model = semantic_segmentation()
model.load_pascalvoc_model(model_file)
```

```python
#모델에 이미지 입력하기
segvalues1, output1 = model.segmentAsPascalvoc(img_path1)
```

#### 3) 이미지 확인하기

```python
# output 이미지 확인하기
plt.imshow(output1)
```

```python
segvalues1
```

사람과 배경뿐만 아니라, diningtable, sofa 가 함께 검출되었다. 여기서 사람만 추출해야겠다.
<br><br>
참고로<br>
LABEL_NAMES = [
    'background', 'aeroplane', 'bicycle', 'bird', 'boat', 'bottle', 'bus',
    'car', 'cat', 'chair', 'cow', 'diningtable', 'dog', 'horse', 'motorbike',
    'person', 'pottedplant', 'sheep', 'sofa', 'train', 'tv'
]

```python
# 물체마다 어떤 색상으로 나타나 있는지 확인한다
# PixelLib에서 나온 코드 그대로 사용

colormap1 = np.zeros((256, 3), dtype=int)
ind = np.arange(256, dtype=int)

for shift in reversed(range(8)):
    for channel in range(3):
        colormap1[:, channel] |= ((ind >> channel) & 1) << shift
    ind >>= 3
    
colormap1[:20]
```

```python
# 사람을 나타내는 15번째 색상
colormap1[15]
```

```python
seg_color1 = (128, 128, 192)

seg_map1 = np.all(output1 == seg_color1, axis=-1)
print(seg_map1.shape)
plt.imshow(seg_map1, cmap='gray')
plt.show()
```

위 사진과 비교하면 사람만 깔끔하게 나왔다.<br>
이제 물체가 있는 위치는 1(True), 그 외에는 0(False)

```python
img_show1 = img_orig1.copy()

#True와 False인 값을 각각 255와 0으로 바꿔준다
img_mask = seg_map1.astype(np.uint8) * 255

color_mask = cv2.applyColorMap(img_mask, cv2.COLORMAP_JET)

img_show1 = cv2.addWeighted(img_show1, 0.6, color_mask, 0.4, 0.0)

plt.imshow(cv2.cvtColor(img_show1, cv2.COLOR_BGR2RGB))
plt.show()
```

#### 4) 배경 흐리게 하기

```python
img_orig1_blur = cv2.blur(img_orig1, (13,13))
plt.imshow(cv2.cvtColor(img_orig1_blur, cv2.COLOR_BGR2RGB))
plt.show()
```

```python
# mask를 이용해 배경만 추출
img_mask_color = cv2.cvtColor(img_mask, cv2.COLOR_GRAY2BGR)
img_bg_mask = cv2.bitwise_not(img_mask_color)
img_bg1_blur= cv2.bitwise_and(img_orig1_blur, img_bg_mask)
plt.imshow(cv2.cvtColor(img_bg1_blur, cv2.COLOR_BGR2RGB))
plt.show()
```

#### 5) 흐린배경과 원본 합성

```python
img1_concat = np.where(img_mask_color==255, img_orig1, img_bg1_blur)
plt.imshow(cv2.cvtColor(img1_concat, cv2.COLOR_BGR2RGB))
plt.show()
```

```python
#사진 저장
cv2.imwrite('../data/human_segmentation/images/blur_jun.jpg', img1_concat)
```

### 1-2. 양 사진

#### 1) 이미지 가져오기

```python
img_path2 = '../data/human_segmentation/images/sheep.jpg'
img_orig2 = cv2.imread(img_path2)

print(img_orig2.shape)
plt.imshow(cv2.cvtColor(img_orig2, cv2.COLOR_BGR2RGB))
plt.show()
```

#### 2) 모델은 위에서 불러온 것 씀

```python
plt.imshow(output2)
```

양과 사람의 팔까지 추출된 것을 볼 수 있다. 여기서 사람은 빼고 양만 추출해야겠다.

```python
# 위에서 뽑은 colormap1에서 
# 양을 나타내는 8번째 색상을 가져온다
colormap1[17]
```

##### seg_color를 설정하고 이것으로만 이루어진 마스크를 만든다.

```python
seg_color2 = (0, 64, 128)

seg_map2 = np.all(output2==seg_color2, axis=-1)
print(seg_map2.shape)
plt.imshow(seg_map2, cmap='gray')
plt.show()
```

원본의 채널정보를 없애고 0과 1로만 이루어진 배열로 바꾸니, 잘 보인다.

```python
img_show2 = img_orig2.copy()

#True와 False 인 값을 각각 255와 0으로 바꿔준다
img_mask = seg_map2.astype(np.uint8) * 255

# 255와 0을 적당한 색상으로 바꿔본다
color_mask = cv2.applyColorMap(img_mask, cv2.COLORMAP_JET)

img_show2 = cv2.addWeighted(img_show2, 0.6, color_mask, 0.4, 0.0)

plt.imshow(cv2.cvtColor(img_show2, cv2.COLOR_BGR2RGB))
plt.show()
```

#### 4) 배경 흐리게 하기

```python
img_orig2_blur = cv2.blur(img_orig2, (13,13))
plt.imshow(cv2.cvtColor(img_orig2_blur, cv2.COLOR_BGR2RGB))
plt.show()
```

```python
# 흐려진 이미지에서 배경만 추출
img_mask_color = cv2.cvtColor(img_mask, cv2.COLOR_GRAY2BGR)
img_bg_mask = cv2.bitwise_not(img_mask_color)
img_bg2_blur = cv2.bitwise_and(img_orig2_blur, img_bg_mask)
plt.imshow(cv2.cvtColor(img_bg2_blur, cv2.COLOR_BGR2RGB))
plt.show()
```

```python
# 이미지 저장
cv2.imwrite('../data/human_segmentation/images/sheep_mask.jpg', img_bg2_blur)
```

#### 5) 흐린배경과 원본 합성

```python
img2_concat = np.where(img_mask_color==255, img_orig2, img_bg2_blur)
plt.imshow(cv2.cvtColor(img2_concat, cv2.COLOR_BGR2RGB))
plt.show()
```

```python
# 이미지 저장
cv2.imwrite('../data/human_segmentation/images/sheep_blur.jpg', img2_concat)
```

양이 여러마리 있어서 가운데에 있는 양 옆의 양도 선명하게 보인다. 원본 사진의 해상도가 낮아졌는지, 블러처리 전과 후의 차이가 극명해보이진 않는다. 사진을 따로 저장하면 잘 보이려나?

### 1-3. 크로마 키 합성하기

#### 1) 이미지 가져오기_합성이미지&배경이미지

```python
# 합성하려는 이미지 읽기
img_path3 = '../data/human_segmentation/images/cat.jpg'
img_orig3 = cv2.imread(img_path3)

print(img_orig3.shape)
plt.imshow(cv2.cvtColor(img_orig3, cv2.COLOR_BGR2RGB))
plt.show()
```

```python
# 배경 이미지 가져오기
bg_path = '../data/human_segmentation/images/back.jpg'
back = cv2.imread(bg_path)

print(back.shape)
plt.imshow(cv2.cvtColor(back, cv2.COLOR_BGR2RGB))
plt.show()
```

```python
# 배경이미지를 합성 이미지에 맞게 resize 한다.
resizedback = cv2.resize(back, (1080, 1440), interpolation = cv2.INTER_CUBIC)

print(resizedback.shape)
plt.imshow(cv2.cvtColor(resizedback, cv2.COLOR_BGR2RGB))
plt.show()
```

#### 2) 모델에 이미지 입력 

```python
segvalues3, output3 = model.segmentAsPascalvoc(img_path3)
```

```python
plt.imshow(output3)
```

```python
segvalues3
```

어두워서 잘 안 보이지만 segvalues를 통해 확인해보니 고양이가 잘 검출된 것을 알 수 있다. 

```python
# 위에서 뽑은 colormap1에서 
# 고양이를 나타내는 8번째 색상을 가져온다
colormap1[8]
```

```python
seg_color3 = (0, 0, 64)

seg_map3 = np.all(output3==seg_color3, axis=-1)
print(seg_map3.shape)
plt.imshow(seg_map3, cmap='gray')
plt.show()
```

원본의 채널 정보를 없애고 0과 1의 배열로 바꾸니 검출된 고양이가 잘 보인다. 

```python
img_show3 = img_orig3.copy()

#True와 False 인 값을 각각 255와 0으로 바꿔준다
img_mask = seg_map3.astype(np.uint8) * 255
plt.imshow(cv2.cvtColor(img_mask, cv2.COLOR_BGR2RGB))
plt.show()

# 255와 0을 적당한 색상으로 바꿔본다
color_mask = cv2.applyColorMap(img_mask, cv2.COLORMAP_JET)
plt.imshow(cv2.cvtColor(color_mask, cv2.COLOR_BGR2RGB))
plt.show()

img_show3 = cv2.addWeighted(img_show3, 0.6, color_mask, 0.4, 0.0)

plt.imshow(cv2.cvtColor(img_show3, cv2.COLOR_BGR2RGB))
plt.show()
```

```python
# 고양이만 남겨두고 검정으로 변형
img_mask_color = cv2.cvtColor(img_mask, cv2.COLOR_GRAY2BGR)
#img_cat_mask = cv2.bitwise_not(img_mask_color)
img_cat = cv2.bitwise_and(img_orig3, img_mask_color)
plt.imshow(cv2.cvtColor(img_cat, cv2.COLOR_BGR2RGB))
plt.show()
```

```python
# 두 이미지를 합성한다.
img_concat_cat = np.where(img_mask_color==255, img_cat, resizedback)
plt.imshow(cv2.cvtColor(img_concat_cat, cv2.COLOR_BGR2RGB))
plt.show()
```

```python
# 이미지 저장
cv2.imwrite('../data/human_segmentation/images/cat_beach.jpg', img_concat_cat)
```

이미지가 잘 합성된 것을 볼 수 있다.
